```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('chicago_taxi_train.csv')
# label_fare = df.corr(numeric_only=True) # ['FARE']
# print(label_fare.sort_values())
sns.pairplot(df, vars=["FARE", "TRIP_MILES"], hue="TRIP_MILES")
```

```python
df = pd.read_csv('Salary_Data.csv') # ['YearsExperience', 'Salary']
# sns.pairplot(df, vars=['YearsExperience', 'Salary'], diag_kind='kde')

yearsOfExp = df['YearsExperience'].values # X
salary = df['Salary'].values # Y

sns.scatterplot(data=salary)
plt.title("Salary data")
plt.xlabel('YearsOfExp')
plt.ylabel('Salary')
```

```python
df = sns.load_dataset('taxis')
print(df.columns.values)
print(df['distance'].head(), df['fare'].head())
sns.pairplot(df, vars=['distance', 'tip', 'fare', 'total'])
```

```python
weight = 0.0
bias = 0.0

# Hyperparameters
learning_rate = 0.01
batch = 5
epochs = 10000

x = yearsOfExp # feature / input
y = salary # label / data_sample

loss = []

for i in range(0, epochs):
    y_pred = weight * x + bias
    mse = (1/batch) * np.sum((y - y_pred) ** 2)

    # Gradient
    dw = (-2/batch) * np.sum(x * (y - y_pred))
    db = (-2/batch) * np.sum(y - y_pred)

    # update parameters
    weight -= learning_rate * dw
    bias -= learning_rate * db

    loss.append(float(mse))

# Dataset
sns.scatterplot(data=salary, label='Dataset')
plt.title("Salary data")
plt.xlabel('YearsOfExp')
plt.ylabel('Salary')

# Regression Line
model = weight * x + bias
plt.plot(x, model, color='red', label='Regression Line')
plt.legend()

print(loss[-5:-1])

# Loss
plt.figure()
plt.plot(loss, label='Training Loss', color='blue')
plt.title('Loss per Iteration')
plt.xlabel('Iteration / Epoch')

plt.ylabel('Loss Value')
```
