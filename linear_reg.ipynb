```python
df = sns.load_dataset('taxis')
# print(df.columns.values)
# sns.pairplot(df, vars=['distance', 'tip', 'fare', 'total'])

# size: 6433
distance = df['distance'].values
fare = df['fare'].values
```

```python
weight = 0.0
bias = 0.0

# Hyperparameters
learning_rate = 0.01
batch = 500
epochs = 10

x = distance # feature / input
y = fare # label / data_sample

loss = []

batch_x = np.array_split(x, (x.__len__() / batch))
batch_y = np.array_split(y, (y.__len__() / batch))

# Axes
figure, (model, error) = plt.subplots(1, 2, figsize=(10, 5))


def plot_graph(x, y, regression, loss):
    model.scatter(x=x, y=y, s=20, label='Dataset')

    model.plot(x, regression, label='Regression Line', color='red')

    error.plot(loss, label='Training Loss', color='blue')

    model.set_title("Salary data")
    model.set_xlabel('YearsOfExp')
    model.set_ylabel('Salary')

    error.set_title("Loss per Iteration")
    error.set_xlabel('Iteration / Epoch')
    error.set_ylabel('Loss Value')

    clear_output(wait=True) # Clear previous output
    display(plt.gcf()) # Display current figure
    # model.clear()
    time.sleep(0.1)

for i in range(0, epochs):

    for i in range(0, batch_x.__len__()):

        x = batch_x[i]
        y = batch_y[i]

        n = x.__len__()

        # Forward pass
        y_pred = weight * x + bias

        # Calculate loss
        mse = (1/n) * np.sum((y - y_pred) ** 2)

        # Gradient
        dw = (-2/n) * np.sum(x * (y - y_pred))
        db = (-2/n) * np.sum(y - y_pred)

        # Update parameters
        weight -= learning_rate * dw
        bias -= learning_rate * db

        loss.append(float(mse))

        plot_graph(x, y, weight * x + bias, loss)


print(loss[-10:-1])


# # Dataset
# sns.scatterplot(x=x, y=y, label='Dataset')
# plt.title("Salary data")
# plt.xlabel('YearsOfExp')
# plt.ylabel('Salary')

# # Regression Line
# model = weight * x + bias
# plt.plot(x, model, color='red', label='Regression Line')
# plt.legend()

# print(loss[-5:-1])

# # Loss
# plt.figure()
# plt.plot(loss, label='Training Loss', color='blue')
# plt.title('Loss per Iteration')
# plt.xlabel('Iteration / Epoch')
# plt.ylabel('Loss Value')
```
